```python
from heapq import heapify,heappush,heappop
import sys
import time
import random

#Create a list of unique numbers from 0 to 8
unique_numbers = list(range(9))
random.shuffle(unique_numbers)

#Reshape the list into a 3x3 matrix
matrix1 = [unique_numbers[i:i+3] for i in range(0, 9, 3)]


initial_state =matrix1
random.shuffle(unique_numbers)
matrix2 = [unique_numbers[i:i+3] for i in range(0, 9, 3)]
goal_state = matrix2



#to travel left , right ,top , bottom
rows = [1, 0, -1, 0]
columns = [0, -1, 0, 1]

initial_time=time.time()

#defining node
class Node:
    def __init__(self, puzzle, parent, heuristic, depth, x, y):
        self.puzzle = puzzle
        self.parent = parent
        self.heuristic =heuristic
        self.depth = depth
        self.x = x
        self.y = y

    def __lt__(self, other):
           return self.heuristic< other.heuristic

    def __eq__(self, other):
           return self.puzzle == other.puzzle


    #checking the boundary conditions of puzzle
    def safe(self, x, y):
        if 0 <= x < 3 and 0 <= y < 3:
            return True
        return False

    #getting heutr
    def get_cost(self):
        count = 0
        for i in range(3):
            for j in range(3):
                if self.puzzle[i][j] != goal_state[i][j]:
                    count += 1
        return count

    def create_child(self, newX, newY):
        new_puzzle = [row[:] for row in self.puzzle]
        new_puzzle[self.x][self.y], new_puzzle[newX][newY] = new_puzzle[newX][newY], new_puzzle[self.x][self.y]
        cost=new_puzzle.get_cost()
        new_node = Node(new_puzzle, self,cost, self.heuristic+ 1, self.depth + 1, newX, newY)
        return new_node

    def goal_reached(self):
        return self.get_cost() == 0

    def check_valid(self):
        count = 0
        temp = []
        for i in range(3):
            for j in range(3):
                temp.append(self.puzzle[i][j])

        #heuristic functions to check boolean
        for i in range(len(temp) - 1):
            for j in range(i + 1, len(temp)):
                if (temp[i] != 0 and temp[j] != 0 and temp[i] > temp[j]):
                    count += 1
        return count % 2 == 0

    def print_mat(self):
        for i in range(3):
            for j in range(3):
                print(self.puzzle[i][j], end=" ")
            print()
        print()

#using recurrsion to trace steps intial state to goal_state
def print_path(node):
    if node is None:
        return
    print_path(node.parent)
    node.print_mat()
    print()

def depth(value):
     return value

def astar(node):
    visited = set()
    min_heap = []
    node.heuristic = node.get_cost()
    heappush(min_heap, (node.heuristic, node))
    #visited.add(tuple(tuple(row) for row in node.puzzle))  # Convert puzzle state to a tuple

    while min_heap:
        temp = heappop(min_heap)[1]
        visited.add(tuple(tuple(row) for row in temp.puzzle))
        if temp.goal_reached():
            print("Goal state reached!")
            print("memory consumed ",  temp.depth*4*sys.getsizeof(temp.puzzle))
            depth(temp.depth)
            print("no of steps==",temp.depth)
            print("total time==",time.time()-initial_time,"sec")


            print_path(temp)
            return True

        x, y = temp.x, temp.y

        for i in range(4):
            newX, newY = x + rows[i], y + columns[i]

            if temp.safe(newX, newY):
                #numOfNodes=1+numOfNodes
                new_puzzle = [row[:] for row in temp.puzzle]
                new_puzzle[x][y], new_puzzle[newX][newY] = new_puzzle[newX][newY], new_puzzle[x][y]
                new_node = Node(new_puzzle, temp, temp.heuristic+1, temp.depth + 1, newX, newY)
                new_node.cost = new_node.get_cost()

                if tuple(tuple(row) for row in new_puzzle) not in visited:

                    heappush(min_heap, (new_node.cost, new_node))
        heapify(min_heap)
    return False


def solve(x, y):
    node = Node(initial_state, None, 0, 0, x, y)



    if astar(node):
        print("Success")
    else:
        print("Goal state not reachable within max depth.")
        inital_function()

        for i in range(3):
            for j in range(3):
               if(initial_state[i][j]==0):
                   solve(i,j)

def inital_function():
    global initial_state
    matrix3=[unique_numbers[i:i+3] for i in range(0, 9, 3)]
    initial_state = matrix3

def goal_function():
    global initial_state
    matrix4=[unique_numbers[i:i+3] for i in range(0, 9, 3)]
    goal_state_state = matrix4

for i in range(3):
    for j in range(3):
        if(initial_state[i][j]==0):
            solve(i,j)
```
