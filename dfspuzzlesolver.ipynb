```python

initial_state = [
    [1, 8, 2],
    [0, 4, 3],
    [7, 6, 5]
]

goal_state = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0]
]

#to travel left , right ,top , bottom
rows = [1, 0, -1, 0]
columns = [0, -1, 0, 1]


#defining node
class Node:
    def __init__(self, puzzle, parent, cost, depth, x, y):
        self.puzzle = puzzle
        self.parent = parent
        self.cost = cost
        self.depth = depth
        self.x = x
        self.y = y

    #checking the boundary conditions of puzzle
    def safe(self, x, y):
        if 0 <= x < 3 and 0 <= y < 3:
            return True
        return False

    #getting cost
    def get_cost(self):
        count = 0
        for i in range(3):
            for j in range(3):
                if self.puzzle[i][j] != goal_state[i][j]:
                    count += 1
        return count

    def create_child(self, newX, newY):
        new_puzzle = [row[:] for row in self.puzzle]
        new_puzzle[self.x][self.y], new_puzzle[newX][newY] = new_puzzle[newX][newY], new_puzzle[self.x][self.y]
        new_node = Node(new_puzzle, self, self.cost + 1, self.depth + 1, newX, newY)
        return new_node

    def goal_reached(self):
        return self.get_cost() == 0

    def check_valid(self):
        count = 0
        temp = []
        for i in range(3):
            for j in range(3):
                temp.append(self.puzzle[i][j])

        #checking condition as hint given in question
        for i in range(len(temp) - 1):
            for j in range(i + 1, len(temp)):
                if (temp[i] != 0 and temp[j] != 0 and temp[i] > temp[j]):
                    count += 1
        return count % 2 == 0

    def print_mat(self):
        for i in range(3):
            for j in range(3):
                print(self.puzzle[i][j], end=" ")
            print()
        print()

#using recurrsion to trace steps intial state to goal_state
def print_path(node):
    if node is None:
        return
    print_path(node.parent)
    node.print_mat()
    print()


def dfs(node, max_depth):
    visited = set()
    stack = []
    stack.append(node)
    visited.add((tuple(row) for row in node.puzzle))

    while stack:
        node = stack.pop(-1)
        if node.goal_reached():
            print("Goal state reached!")
            print_path(node)
            return True
        if node.depth >= max_depth:
            continue
        x, y = node.x, node.y

        for i in range(4):
            newX, newY = x + rows[i], y + columns[i]

            if node.safe(newX, newY):
                new_puzzle = [row[:] for row in node.puzzle]
                new_puzzle[x][y], new_puzzle[newX][newY] = new_puzzle[newX][newY], new_puzzle[x][y]
                new_node = Node(new_puzzle, node, node.cost + 1, node.depth + 1, newX, newY)
                new_node.cost = new_node.get_cost()

                if (tuple(row) for row in new_puzzle) not in visited:
                    visited.add((tuple(row) for row in new_puzzle))
                    stack.append(new_node)
    return False

def solve(x, y, max_depth):
    node = Node(initial_state, None, 0, 0, x, y)

    if not node.check_valid():
        print("Goal state not reachable.")
        return

    if dfs(node, max_depth):
        print("Success")
    else:
        print("Goal state not reachable within max depth.")

solve(1, 0, 12)
```
