```python
import time
t1=time.time()
recurssions=0

initial_state = [
    [5, 3, 0],
    [8, 7, 6],
    [2, 4, 1]
]

goal_state = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0]
]

rows = [1, 0, -1, 0]
columns = [0, -1, 0, 1]

class Node:
    def __init__(self, puzzle, parent=None, heuristic=0, depth=0, x=0, y=2):
        self.puzzle = puzzle
        self.parent = parent
        self.heuristic = heuristic
        self.depth = depth
        self.x = x
        self.y = y
        self.visited = set()  # Track visited states

    def safe(self, x, y):
        if 0 <= x < 3 and 0 <= y < 3:
            return True
        return False

    def get_cost(self):
        count = 0
        for i in range(3):
            for j in range(3):
                if self.puzzle[i][j] != goal_state[i][j]:
                    count += 1
        return count

    def create_child(self, newX, newY):
        new_puzzle = [row[:] for row in self.puzzle]
        new_puzzle[self.x][self.y], new_puzzle[newX][newY] = new_puzzle[newX][newY], new_puzzle[self.x][self.y]
        new_node = Node(new_puzzle, self, 0, self.depth + 1, newX, newY)
        new_node.heuristic = new_node.get_cost()
        return new_node

    def goal_reached(self):
        return self.get_cost() == 0

    def print_mat(self):
        for i in range(3):
            for j in range(3):
                print(self.puzzle[i][j], end=" ")
            print()
        print()

def print_path(node):
    if node is None:
        return
    print_path(node.parent)
    node.print_mat()
    print()

def rbfs(node, f_limit):
    global recurssions
    if node.goal_reached():
        return node, 0

    successors = []

    for i in range(4):
        newX, newY = node.x + rows[i], node.y + columns[i]

        if node.safe(newX, newY):
            new_node = node.create_child(newX, newY)
            if tuple(map(tuple, new_node.puzzle)) not in node.visited:  # Check if state is not visited
                successors.append(new_node)
                new_node.visited = node.visited.copy()  # Copy parent's visited set
                new_node.visited.add(tuple(map(tuple, new_node.puzzle)))  # Add current state

    if not successors:
        return None, 500

    while True:
        successors.sort(key=lambda x: x.heuristic + x.depth)

        best = successors[0]

        if best.heuristic + best.depth > f_limit:
            return None, best.heuristic + best.depth

        alternative = successors[1].heuristic + successors[1].depth if len(successors) > 1 else 500
        result, best.heuristic = rbfs(best, min(f_limit, alternative))
        recurssions+=1

        if result is not None:
            return result, 0

def recursive_best_first_search(initial_state):
    node = Node(initial_state)
    return rbfs(node,500)

result, _ = recursive_best_first_search(initial_state)


if result is not None:
    print("Solution found:")
    print_path(result)
else:
    print("No solution found.")

print("time=",time.time()-t1)
print("no of recurssions=",recurssions)
```
