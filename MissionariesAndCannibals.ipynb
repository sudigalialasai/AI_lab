```python
# Defining basic node structure
#index at 0 is no of Missionaries on left side
#index at 1 is no of Cannibals on right side
#index at 2 is position of river(1->left side and 2->right side)
initial_state = [3,3,1]

#finally there are zero number of Missionaries and Cannibals on left side
goal_state = [0,0,0]



operations=[[1,0,1] #only one Missionaries moving on ship
            ,[0,1,1] #only one Cannibals moving on ship
            ,[2,0,1] #only two Missionaries moving on ship
            ,[0,2,1] #only two Cannibals moving on ship
            ,[1,1,1] #only one Missionaries and one Cannibals moving on ship
            ]

class Node:
    def __init__(self, puzzle, parent,depth):
        self.puzzle = puzzle
        self.parent = parent
        self.depth = depth  #depth is nothing but cost of each child


    def safe(self, x, y,z):
        if x==0:
             return True
        if 0 <= x <= 3 and 0 <= y <=3 and  x>=y:
           right_x=3-x
           right_y=3-y
           if right_x==0:
                   return True
           return 3-x>=3-y
        return False

    def create_child(self, newX, newY,newZ):
        new_puzzle =[newX, newY,newZ]
        new_node = Node(new_puzzle, self, self.depth + 1)
        return new_node

    def goal_reached(self):
        return self.puzzle[0]==0 and self.puzzle[1]==0 and self.puzzle[2]==0


    def print_mat(self):

                if self.puzzle[2]==1:
                      print(self.puzzle[0],"  missionaries and ",self.puzzle[1] ," cannibals are their and ship at left side")
                else:
                      print(self.puzzle[0],"  missionaries and ",self.puzzle[1] ," cannibals are their and ship at right side")

                print()

def print_path(node):
    if node is None:
        return
    print_path(node.parent)
    node.print_mat()
    print()

def bfs(initial_node):
    queue = []
    queue.append(initial_node)

    while queue:
        node = queue.pop(0)
        if node.goal_reached():
            print("Goal state reached!")
            print_path(node)
            return True


        for i in range(5):
          if(node.puzzle[2]==1):
            newX, newY,newZ =node.puzzle[0]-operations[i][0],node.puzzle[1]-operations[i][1],node.puzzle[2]-operations[i][2]
          else:
             newX, newY,newZ =node.puzzle[0]+operations[i][0],node.puzzle[1]+operations[i][1],node.puzzle[2]+operations[i][2]

          if node.safe(newX, newY,newZ):
               queue.append(node.create_child(newX, newY,newZ))
    return False

def solve():
    node = Node(initial_state, None, 0)


    if bfs(node):
        print("Success")
    else:
        print("Goal state not reachable within max depth.")

solve()
```
